Assumptions:

If corr in range 0.8 to 1 - Highly related
If corr in range 0.6 to 0.8 - Moderately related
If corr in range 0.4 to 0.6 - Weakly related
If corr in range 0 to 0.4 - Negligible/Unrelated

If corr is +ve then relationship is proportionally affected (as one increases, other increases and vice-versa)
If corr is -ve then relationship is negatively affected (as one increases, other decreases and vice-versa)

Below are the inferences we can draw based on the above data and assumption from the linear correlation matrix:

    1. Signal2 and Signal6 are the signals that affect the Price the most (positively), i.e. as Signal2 and/or Signal6 increases, Price increases and vice versa.
    2. For price to increase Signal3 and Signal4 have to decrease while Signal1, Signal2, Signal6 and Signal7 have to increase and vice versa. But all these signals are co-related with each other.
    
Below lets analyze the Highly Related correlation alone i.e. 0.8 to 1:

    3. Signal3 and Signal4 are related with a positive correlation, which means that they reinforce each other, i.e. as Signal3 decreases Signal4 also decreases and hence increasing Price, and vice versa.
    4. There is a also a strong correlation between Signal6 and Signal7 suggesting that they reinforce each other, i.e. as Signal6 increases Signal7 also increases and hence increasing Price, and vice versa.
    5. Then there is a box of correlations which imply that Signal6 and Signal7 are negatively correlated to Signal3 and Signal4. These 4 correlations (i.e. 3&6, 3&7, 4&6, 4&7) suggest that if Signal3 and Signal4 decrease then Signal6 and Signal7 increase and both imply that Price increases, and vice versa.
   From the above 3 point we can infer that Signal6 and Signal3 directly affect Price while Signal4 and Signal7 do not affect Price significantly but do affect it indirectly. Signal4 and Signal7 are positively related (reinforces) with Signal3 and Signal6 respectively. And then pair Signal3 and Signal4 are negatively related to Signal6 and Signal7 pair, and the former pair being negatively related to Price while the latter pair being positively related to Price. 
   Hence a decrease in pair Signal3 and Signal4 and an increase in pair Signal6 and Signal7 leads to an increase in Price without objections and vice versa.

Below lets analyze the Moderately Related correlation alone i.e. 0.6 to 0.8:

    6. Signal2 and Signal6 are positively related with Price. An increase in them increases the Price variable.
    7. Signal1 and Signal2 are positively related with each other. An increase in Signal1 leads to increae in Price in 2 steps and hence its effect is low.
    8. Signal4 and Signal1 are positively related with each other. Since we've established that Signal4 has to decrease hence price will decrease on a 3 step process whose effect is almost inefficient. That is if Signal4 decreases it decreases Signal1 which in turns decreases Signal2 which in turns decreases the Price which is very long process with only moderate relationships, hence this path can be discarded.
   From the above 3 points we can infer that Signal2 and Signal6 have to increase for Price to increase and vice versa. Signal1's situation is a bit tricky. On one hand it has to increase to strenghten the increase with Signal2 and hence increase the price. On the other hand if it increases it affects the extent Signal4 can decrease. But Signal1 has to increase even because as we'll infer in the next point, it has a direct weak relation with price. So Signal1 has to increase but not too much wherein the combined increase in price due to the relation directly with price and the relation with Signal2 is outweighed by the decrease in Price due to its relation with Signal4. So Signal1 has to increase but for every combination of increase in Signals 2,6,7 and decrease in Signals 3,4 there will be an ideal increase in Signal1 for maximum increase in Price.The same applies if price has to decrease.
   
Below lets analyze the Weakly Related correlation alone i.e. 0.4 to 0.6:

    9. Signal1 is positively related and Signal3 is negatively related with Price.
    10. Other weak relations don't take any effect as they are 2 or more steps away from affecting Price and hence have negligible effect.
    
   RESULT: From all these inferences we can infer that for maximum increase in Price; Signals 2,6,7 have to increase and Signals 3,4 have to decrease, and for the combinations of these 5 signals, Price increases as Signal1 increases upto a specific point, beyond which Price decreases from its maximum increase Price point as Signal1 increases further.

MODEL 1: K-NEAREST NEIGHBORS (1-NEIGHBOR)

```python
df1 = df_critical.drop('Price',axis=1)
ds1 = df1.to_numpy()
print(ds1)
```

```python
df2 = df_critical['Price']
ds2 = df2.to_numpy()
ds2 = ds2.flatten()
print(ds2)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(ds1,ds2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn1 = KNeighborsRegressor(n_neighbors=1)
knn1.fit(X_train, y_train)
print('Training Set Score: {}'.format(knn1.score(X_train,y_train)))
print('Test Set Score: {}'.format(knn1.score(X_test,y_test)))
```

MODEL 2: K-NEAREST NEIGHBORS (N-NEIGHBORS)

```python
training_accuracy = []
test_accuracy = []
neighbors_settings = range(1,11)
for n in neighbors_settings:
    clf = KNeighborsRegressor(n_neighbors=n)
    clf.fit(X_train,y_train)
    training_accuracy.append(clf.score(X_train,y_train))
    test_accuracy.append(clf.score(X_test,y_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbors_settings, training_accuracy, label='Training Accuracy')
plt.plot(neighbors_settings, test_accuracy, label='Test Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('n_neighbors')
plt.legend()
```

Maximum test accuracy is for 3 neighbors

```python
knn2 = KNeighborsRegressor(n_neighbors=3)
knn2.fit(X_train, y_train)
print('Training Set Score: {}'.format(knn2.score(X_train,y_train)))
print('Test Set Score: {}'.format(knn2.score(X_test,y_test)))
```

MODEL 3: LINEAR REGRESSION (ordinary least squares)

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train,y_train)
print('w: {}'.format(lr.coef_))
print('b: {}'.format(lr.intercept_))
print('Training Set Score: {}'.format(lr.score(X_train,y_train)))
print('Test Set Score: {}'.format(lr.score(X_test,y_test)))
```

MODEL 4: RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train,y_train) # alpha = 1, alpha is learning rate
print('Training Set Score: {}'.format(ridge.score(X_train,y_train)))
print('Test Set Score: {}'.format(ridge.score(X_test,y_test)))
```

```python
ridge10 = Ridge(alpha=10).fit(X_train,y_train)
print('Training Set Score: {}'.format(ridge10.score(X_train,y_train)))
print('Test Set Score: {}'.format(ridge10.score(X_test,y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train,y_train)
print('Training Set Score: {}'.format(ridge01.score(X_train,y_train)))
print('Test Set Score: {}'.format(ridge01.score(X_test,y_test)))
```

```python
ridge001 = Ridge(alpha=0.01).fit(X_train,y_train)
print('Training Set Score: {}'.format(ridge001.score(X_train,y_train)))
print('Test Set Score: {}'.format(ridge001.score(X_test,y_test)))
```

As we can see from the above example as we decrease alpha (regularization parameter), accuracy increases but the rate of increase of accuracy drops sharply. Linear regression is ridge with alpha = 0.

MODEL 5: DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(random_state=0).fit(X_train,y_train)
print('Training Set Score: {}'.format(tree.score(X_train,y_train)))
print('Test Set Score: {}'.format(tree.score(X_test,y_test)))
```

Since training set score is 1 it means it is likely overfitting. To reduce over fitting we apply stronger pre-pruning by limiting the maximum depth.

```python
training_accuracy = []
test_accuracy = []
depth = range(10,25)
for i in depth:
    tree= DecisionTreeRegressor(max_depth=i, random_state=0)
    tree.fit(X_train,y_train)
    training_accuracy.append(tree.score(X_train,y_train))
    test_accuracy.append(tree.score(X_test,y_test))
```

```python
plt.plot(depth, training_accuracy, label='Training Accuracy')
plt.plot(depth, test_accuracy, label='Test Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Maximum depth')
plt.legend()
```

At max_depth = 13, the test score peaks and train score is not 1 simultaneously.

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(random_state=0,max_depth=13).fit(X_train,y_train)
print('Training Set Score: {}'.format(tree.score(X_train,y_train)))
print('Test Set Score: {}'.format(tree.score(X_test,y_test)))
```

MODEL 6: RANDOM FORESTS (with 1 estimator)

```python
from sklearn.ensemble import RandomForestRegressor
forest1 = RandomForestRegressor(n_estimators=1, random_state=0)
forest1.fit(X_train,y_train)
print('Training Set Score: {}'.format(forest1.score(X_train,y_train)))
print('Test Set Score: {}'.format(forest1.score(X_test,y_test)))
```

MODEL 7: RANDOM FORESTS (with n estimators)

```python
training_accuracy = []
test_accuracy = []
estimators = range(1,100,10)
for i in estimators:
    random = RandomForestRegressor(n_estimators=i, random_state=0)
    random.fit(X_train,y_train)
    training_accuracy.append(random.score(X_train,y_train))
    test_accuracy.append(random.score(X_test,y_test))
```

```python
plt.plot(estimators, training_accuracy, label='Training Accuracy')
plt.plot(estimators, test_accuracy, label='Test Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('No. of Estimators')
plt.legend()
```

After 20 there is no significant increase in accuracy

```python
from sklearn.ensemble import RandomForestRegressor
forest1 = RandomForestRegressor(n_estimators=20,random_state=0)
forest1.fit(X_train,y_train)
print('Training Set Score: {}'.format(forest1.score(X_train,y_train)))
print('Test Set Score: {}'.format(forest1.score(X_test,y_test)))
```
